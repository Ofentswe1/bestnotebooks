**Steganography** - is the practice of concealing a file, message, image, or video within another file,
message, image, or video. The word steganography combines the Greek words steganos (στεγανός), meaning 
"covered, concealed, or protected," and graphein (γράφειν) meaning "writing".

images as represented by red, green, blue (RGB) scale in different color code. The notebook covers the below example of this image.

Image of a tree with a steganographically hidden image. The hidden image is revealed by removing all but the two least significant bits of each color component and a subsequent normalization. The hidden image is shown below.

## Load packages  
___ 

we are using PIL, numpy and matplotlib to load images, convert images to arrays and plot images

```python
import numpy as np 
import matplotlib.pyplot as plt
from PIL import Image
%matplotlib inline
```

Load the image from directory images

```python
img = Image.open('../images/steganography.png')
```

Display original image for steganography

```python
img
```

Covert image to numpy array 

```python
X = np.asarray(img)
```

Display the shape/dimension of the array

```python
X.shape
```

Assigning the length, width and color shape from **X.shape**  

```python
[length, width, rgb] = X.shape
```

1. Setting the mask to the largest two digits bits
2. Setting the image to zeros for decoding 
3. Set make up color to 6 make it darker a bit better

```python
mask = 3
image_dec = np.zeros((length, width, rgb))
makeup_color = 6
```

Decoding the image with numpy bitwise_and operator to reveal the hidden image

```python
for row in range(length):
    for column in range(width):
        for color_type in range(rgb):
            new_image = X[row, column, color_type]
            matrix_bit = np.bitwise_and(new_image, mask)
            image_dec[row, column, color_type] = (makeup_color * makeup_color + makeup_color) * matrix_bit
```

### Showing the decoded image using matplotlib 
___

> “Big Brother is Watching You.”  
> <cite> ― George Orwell, 1984 </cite>

```python
plt.imshow(image_dec/np.max(image_dec))
```
